```python
import pandas as pd
import numpy as np
import random
```

```python
df_raw = pd.read_excel("data/raw/hosong.xlsx", header = [0, 1, 2, 3])
print(f"length: {len(df_raw)}")
print(f"columns: {df_raw.shape[1]}")
df_raw.head(3)
```

```python
# Expand data to 1200
df = pd.DataFrame(columns = df_raw.columns, index = range(1200))
start = len(df_raw)
df.iloc[:start] = df_raw
print(f"length: {len(df)}")
df.head(3)
```

```python
# Generate time
start_date = "2023-09-21 17:36:00"
end_date = "2023-10-6 23:59:59"
time_range = pd.date_range(start = start_date, end = end_date, freq = 'T') # 每5分鐘產生一個時間
random_time = np.random.choice(time_range, size = (len(df) - start))
sort_time = pd.DatetimeIndex(random_time).sort_values()
df.iloc[start:, 0] = sort_time
```

```python
# Get the full names of features of different categories
NV_col = [i for i in df.columns if i[3] == "NV"]
AV_col = [i for i in df.columns if i[3] == "AV"]
DV_col = [i for i in df.columns if i[3] == "DV"]
UT_col = [i for i in df.columns if i[3] == "UT"]
LT_col = [i for i in df.columns if i[3] == "LT"]
ER_col = [i for i in df.columns if i[3] == "ER"]
print(f"NV_col: {len(NV_col)}")
```

```python
# fill constant value
df[NV_col + UT_col + LT_col] = df[NV_col + UT_col + LT_col].fillna(method = "ffill")
df.tail()
```

```python
limit_dict = {
    "直徑(1)": [0.022,  0],
    "Y(7)":    [0.012, -0.01],
    "Z(7)":    [0.01,  -0.01],
    "直徑(2)": [0.022,  0],
    "Y(8)":    [0.01,  -0.01],
    "Z(8)":    [0.01,  -0.015],
    "Y(5)":    [0.019,  0],
    "直徑(3)": [0.022,  0],
    "Y(9)":    [0.01,  -0.013],
    "Z(9)":    [0.01,  -0.011],
    "直徑(4)": [0.024,  0],
    "Y(10)":   [0.012, -0.01],
    "Z(10)":   [0.01,  -0.01],
    "Y(6)":    [0.019, -0.008],    
    "XY(1)":   [0.05,  -0.10],

}
limit = list(limit_dict.values())
```

```python
# Generate AV
for i, (nv_col, av_col) in enumerate(zip(NV_col, AV_col)):
    for j in range(start, len(df)):
        diff1 = round(random.uniform(limit[i][1], 0), 3)
        diff2 = round(random.uniform(0.001, limit[i][0]), 3)
        diff1 = diff2 if (diff1 == 0) else diff1
        new = df.loc[j, nv_col] + random.choice([diff1, diff2])
        new = round(new, 3)
        df.loc[j, av_col] = new
```

```python
df.tail()
```

```python
# Generate DV
df[DV_col] = (df[AV_col].values - df[NV_col].values).astype(float).round(3)
```

```python
# Generate ER
ER_upper_limit = (df[DV_col].values - df[UT_col].values)
ER_lower_limit = (df[DV_col].values - df[LT_col].values)
ER_upper_limit = np.maximum(ER_upper_limit, 0)
ER_lower_limit = np.minimum(ER_lower_limit, 0)
df[ER_col] = (ER_upper_limit + ER_lower_limit).astype(float).round(3)
df.tail()
```

```python
# adjust ER count to achieve ~ 30%
df_ER = df[(df[ER_col] != 0).any(axis = 1)]
adjust_n = round(((len(df_ER)/len(df)) - 0.33) * len(df))
adjust_index = df_ER.iloc[start:].sample(n = adjust_n).index
adjust_index
```

```python
# Generate AV
LT_value = df.loc[0, LT_col]
UT_value = df.loc[0, UT_col]
for i, (nv_col, av_col) in enumerate(zip(NV_col, AV_col)):
    for j in adjust_index:
        diff1 = round(random.uniform(LT_value[i], 0), 3)
        diff2 = round(random.uniform(0.001, UT_value[i]), 3)
        diff1 = diff2 if (diff1 == 0) else diff1
        new = df.loc[j, nv_col] + random.choice([diff1, diff2])
        new = round(new, 3)
        df.loc[j, av_col] = new
```

```python
# Generate DV
df[DV_col] = (df[AV_col].values - df[NV_col].values).astype(float).round(3)
```

```python
# Generate ER
ER_upper_limit = (df[DV_col].values - df[UT_col].values)
ER_lower_limit = (df[DV_col].values - df[LT_col].values)
ER_upper_limit = np.maximum(ER_upper_limit, 0)
ER_lower_limit = np.minimum(ER_lower_limit, 0)
df[ER_col] = (ER_upper_limit + ER_lower_limit).astype(float).round(3)
df.tail()
```
